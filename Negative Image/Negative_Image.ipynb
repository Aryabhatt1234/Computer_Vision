```python
import cv2
import matplotlib.pyplot as plt
```

```python
img_path = 'lake.jpg'
img = cv2.imread(img_path)
#img
```

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
# cv2.imshow('original',img)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
```

```python
gray = cv2.imread(img_path,0)
#gray
# cv2.imshow('gray',gray)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
```

```python
colored_negative = abs(255-img)
gray_negative = abs(255-gray)
```

```python
# cv2.imshow('colored_negative',colored_negative)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
```

```python
# cv2.imshow('gray_negative',gray_negative)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
```

```python
title = ['original','gray', 'colored_negative', 'gray_negative']
imgs = [img,gray,colored_negative,gray_negative]
```

```python
plt.subplot(2,2,1)
plt.title(title[0])
plt.imshow(imgs[0])
plt.xticks([])
plt.yticks([])

plt.subplot(2,2,2)
plt.title(title[1])
plt.imshow(imgs[1])
plt.xticks([])
plt.yticks([])

plt.subplot(2,2,3)
plt.title(title[2])
plt.imshow(imgs[2])
plt.xticks([])
plt.yticks([])

plt.subplot(2,2,4)
plt.title(title[3])
plt.imshow(imgs[3])
plt.xticks([])
plt.yticks([])

plt.show()
```
